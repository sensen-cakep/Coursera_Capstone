```python
#Wikipedia scraping for Toronto postal codes
import pandas as pd
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
df_scrape = pd.read_html(url)
```

```python
#check number of tables found
len(df_scrape)
```

```python
#show table
df_scrape[0].head()
```

```python
#create separate dataframe
df_post_code = df_scrape[0]
df_post_code.head()
```

```python
#remove rows with borough Not assigned

# Get names of indexes
indexNames = df_post_code[ df_post_code['Borough'] == 'Not assigned' ].index

# Delete these row indexes from dataFrame
df_post_code.drop(indexNames , inplace=True)
df_post_code.head()
```

```python
# More than one neighborhood can exist in one postal code area, combined these into one row with the neighborhoods separated with a comma
df_post_code["Neighbourhood"] = df_post_code.groupby("Postal Code")["Neighbourhood"].transform(lambda neigh: ', '.join(neigh))

#remove duplicates
df_post_code = df_post_code.drop_duplicates()

#update index to be Postal Code if it isn't already
#if(df_post_code.index.name != 'Postal Code'):
#    df_post_code = df_post_code.set_index('Postal Code')
    
df_post_code.head()
```

```python
#if a cell has a borough but a Not assigned  neighborhood, then the neighborhood will be the same as the borough
indexNames = df_post_code['Neighbourhood'] == 'Not assigned'
df_post_code['Neighbourhood'][indexNames] = df_post_code['Borough'][indexNames]
df_post_code.head()
```
